```python
# Problem 1.1
Fp,xp = wavfile.read('piano4.wav')
Ft,xt = wavfile.read('trumpet4.wav')
Fv,xv = wavfile.read('violin4.wav')
Ff,xf = wavfile.read('flute4.wav')

plt.subplot(411)
plt.title('Magnitide spectrum of Piano')
xpm = plt.magnitude_spectrum(xp,Fs=Fp)
xpmm = xpm[0]
pm = sp.find_peaks(xpmm,threshold=2,distance=1)
p_f = pm[0][0]

plt.subplot(412)
plt.title('Magnitide spectrum of trumpet')
xtm = plt.magnitude_spectrum(xt,Fs=Ft)
xtmm = xtm[0]
tm = sp.find_peaks(xtmm,threshold=5,distance=1)
t_f = tm[0][0]

plt.subplot(413)
plt.title('Magnitide spectrum of violin')
xvm = plt.magnitude_spectrum(xv,Fs=Fv)
xvmm = xvm[0]
vm = sp.find_peaks(xvmm,threshold=14,distance=2)
v_f = vm[0][0]

plt.subplot(414)
plt.title('Magnitide spectrum of flute')
xfm = plt.magnitude_spectrum(xf,Fs=Ff)
xfmm = xfm[0]
fm = sp.find_peaks(xfmm,threshold=12,distance=2)
f_f = fm[0][0]
plt.subplots_adjust(left=0.1,
                    bottom=-0.5, 
                    right=0.9, 
                    top=0.9, 
                    wspace=0.4, 
                    hspace=0.4)
print("Frequency corresponding to fundamental harmonic for piano is",p_f,"Hz")
print("Frequency corresponding to fundamental harmonic for trumpet is",t_f,"Hz")
print("Frequency corresponding to fundamental harmonic for violin is",v_f,"Hz")
print("Frequency corresponding to fundamental harmonic for flute is",f_f,"Hz")
```

```python
# Problem 1.2
Ff1,xf1 = wavfile.read('flute1.wav')
Ff2,xf2 = wavfile.read('flute2.wav')
Ff3,xf3 = wavfile.read('flute3.wav')

plt.subplot(311)
plt.title('Magnitide spectrum of input flute1.wav')
xf1m = plt.magnitude_spectrum(xf1,Fs=Ff1)
xf1mm = xf1m[0]
f1m = sp.find_peaks(xf1mm,threshold=2,distance=1)
f_f1 = f1m[0][0]

plt.subplot(312)
plt.title('Magnitide spectrum of flute2.wav')
xf2m = plt.magnitude_spectrum(xf2,Fs=Ff2)
xf2mm = xf2m[0]
f2m = sp.find_peaks(xf2mm,threshold=2,distance=1)
f_f2 = f2m[0][0]

plt.subplot(313)
plt.title('Magnitide spectrum of flute3.wav')
xf3m = plt.magnitude_spectrum(xf3,Fs=Ff3)
xf3mm = xf3m[0]
f3m = sp.find_peaks(xf3mm,threshold=2,distance=1)
f_f3 = f3m[0][0]
plt.subplots_adjust(left=0.1,
                    bottom=-0.2, 
                    right=0.9, 
                    top=0.9, 
                    wspace=0.4, 
                    hspace=0.4)
print("Frequency corresponding to fundamental harmonic for flute1.wav is",f_f1,"Hz")
print("Frequency corresponding to fundamental harmonic for flute2.wav is",f_f2,"Hz")
print("Frequency corresponding to fundamental harmonic for flute3.wav is",f_f3,"Hz")
```

```python
# problem 2
Fw,xw = wavfile.read('Whistle.wav')
xw=xw[:,0]
Fin,xin = wavfile.read('R1.wav')

if(len(xin.shape)>1):
    xin=xin[:,0]
    plt.tight_layout()
plt.subplot(211)
plt.title('Magnitide spectrum of keylock')
xwm = plt.magnitude_spectrum(xw,Fs=Fw)
xwmm = xwm[0]
wm = sp.find_peaks(xwmm,threshold=2,distance=1)
f_w = wm[0][0]

def keylock(x):
    plt.subplot(212)
    plt.title('Magnitide spectrum of input signal')
    xcm = plt.magnitude_spectrum(x,Fs=Fin)
    xcmm = xcm[0]
    cm = sp.find_peaks(xcmm,threshold=2,distance=1)
    f_fc = cm[0][0]
    print("Input frequency of fundamental harmonic:",f_fc)
    if(f_fc<=f_w*1.05 and f_fc>=f_w*0.95):
        return 1
print("Keylock:",f_w)
if(keylock(xin)):
    print("ACCESS GRANTED")
else:
    print("ACCESS DENIED")   
plt.subplots_adjust(left=0.1,
                    bottom=0.1, 
                    right=0.9, 
                    top=0.9, 
                    wspace=0.4, 
                    hspace=0.4)
```
