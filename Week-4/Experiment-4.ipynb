```python
# Problem 1.1,1.2
Fs=120
f= 15*alpha
t=np.arange(0,2,1/Fs)
x = np.sin(2*np.pi*f*t)
def DFT(x):
    """
    Function to calculate the 
    discrete Fourier Transform 
    of a 1D real-valued signal x
    """

    N = len(x)
    n = np.arange(N)
    k = n.reshape((N, 1))
    e = np.exp(-2j * np.pi * k * n / N)
    
    X = np.dot(e, x)
    
    return X
X = DFT(x[0:120])
X1 = DFT(x[0:130])
# calculate the frequency
N = len(X)
n = np.arange(N)
T = N/Fs
freq = n/T 

N1 = len(X1)
n1 = np.arange(N1)
T1 = N1/Fs
freq1 = n1/T1

plt.figure(figsize = (8, 6))
plt.stem(freq, abs(X), 'b', \
         markerfmt=" ", basefmt="-b")
plt.xlabel('Freq (Hz)')
plt.ylabel('DFT Amplitude |X(freq)|')
plt.show()
```

```python
plt.figure(figsize = (8, 6))
plt.stem(freq, abs(X), 'b', \
         markerfmt=" ", basefmt="-b")
plt.stem(freq1, abs(X1), 'r', \
         markerfmt=" ", basefmt="-r")
plt.xlabel('Freq (Hz)')
plt.ylabel('DFT Amplitude |X(freq)|')
plt.show()
```
